```python
def to_num(xs):
    new_xs = []
    for i, x in enumerate(xs):
        if type(x) == int:
            new_xs.append(x)
            
        elif x.isnumeric():
            new_xs.append(int(x))
        else:
            _x = x.replace('[', '')
            _x = _x.replace(']', '')
            new_xs.append(int(_x))
    return new_xs
```

# Distance measure

```python
from math import sin, cos, sqrt, atan2, radians 

def distance(city1, city2):
    R = 6373.0
    lat1 = radians(city1.geo_lat)
    lon1 = radians(city1.geo_lon)
    lat2 = radians(city2.geo_lat)
    lon2 = radians(city2.geo_lon)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
def path_distance(cities: [pd.Series]):
    # total distance for path
    d = 0.0
    for i in range(len(cities)-1):
        d += distance(cities[i], cities[i+1])
    return d 
```

# Simulated Annealing

```python
class SimulatedAnnealing:
    def __init__(self, T, max_t, min_T, cities):
        self.T = T
        self.t = 0
        self.max_t = max_t
        self.min_T = min_T
        self.x_t = [i for i in range(len(cities))] # initial order
        self.x_prime = self.x_t
        self.generate()
        self.cities = cities
        self.energy_xt = self.energy(self._in_order_(cities, self.x_t)) # initial energy
        self.alpha = 0.80 
        self.best_position = (self.energy_xt, self.x_t)
        self.energy_prime = self.energy_xt
        
    def energy(self, cities):
        return path_distance(cities)
    def accept(self, x_prime):
        alpha = np.random.uniform(0, 1)
        if self.__prob(self.x_t) == 0.0:
            return True
        ratio = self.__prob(x_prime)/ self.__prob(self.x_t)
        if ratio >= alpha:
            return True
        return False
    def step(self):
        self.energy_prime = self.energy(self._in_order_(self.cities, self.x_prime))
        self.energy_xt = self.energy(self._in_order_(self.cities, self.x_prime))
        if self.accept(self.x_prime):
            self.x_t = self.x_prime
            self.t += 1
            energy_x = self.energy(self._in_order_(self.cities, self.x_prime))
            if energy_x < self.best_position[0]:
                self.best_position = (energy_x, self.x_prime)
            if self.t > self.max_t:
                self.T *= self.alpha
                self.t = 0
        self.generate() # new x_prime
        
        
    def extract_parameters(self, model):
        array = np.concatenate([p.data.flatten() for p in model.parameters()])
        return array
    
    def __prob(self, x_prime):
        cities = self._in_order_(self.cities, x_prime)
        return (np.e)**(-self.energy(cities)/self.T)
    
    def _in_order_(self, cities, indexes):
        new_order = []
        for i in indexes:
            new_order.append(cities[i])
        return new_order
    
    def generate(self):
        # generates new order of cities
        a = np.random.randint(0, len(cities)-1)
        b = np.random.randint(0, len(cities)-1)
        x_prime = self.x_t.copy()
        x_prime[a] = self.x_t[b]
        x_prime[b] = self.x_t[a]
        self.x_prime = x_prime
        
    def get_cities_coords(self, order):
        x = []
        y = []
        names = []
        for i in order:
            city = self.cities[i]
            x.append(city.geo_lon)
            y.append(city.geo_lat)
            names.append(city.address)
        return x, y, names
            
        
```

```python
cities = [top_30.iloc[i] for i in range(len(top_30))] # transfer to [pd.Series]
sa_opt = SimulatedAnnealing(90, 50, 1, cities)
```

```python
# TRAINING + RECORDING DATA
history_x = []
history_y = []
history_names = []
history_d = []
for i in range(3000):
    if i % 30 == 0:
        x, y, names = sa_opt.get_cities_coords(sa_opt.x_t)
        history_x.append(x)
        history_y.append(y)
        history_names.append(names)
        history_d.append(sa_opt.energy_xt)
        print(f'Min dist: {sa_opt.best_position[0]}, T:{sa_opt.T}')
    sa_opt.step()
    
    
x, y, names = sa_opt.get_cities_coords(sa_opt.best_position[1])
history_x.append(x)
history_y.append(y)
history_names.append(names)
history_d.append(sa_opt.best_position[0])
```
